# Comment Extractor

```python
import time
import json
from googleapiclient.errors import HttpError

def extract_comments_and_replies_final(youtube, playlist_ids, max_requests_per_minute=50):
    data = []

    for playlist_id in playlist_ids:
        videos = []

        playlist_request = youtube.playlistItems().list(
            part="snippet",
            playlistId=playlist_id,
            maxResults=50  # we can adjust this value our requirement
        )

        playlist_response = playlist_request.execute()

        for item in playlist_response["items"]:
            video_id = item["snippet"]["resourceId"]["videoId"]
            videos.append(video_id)

        requests_made = 0
        for video_id in videos:
            if requests_made >= max_requests_per_minute:
                # Pause execution to avoid exceeding quota limits
                time.sleep(60)  # Sleep for 60 seconds (1 minute)
                requests_made = 0

            try:
                video_comments = retrieve_all_comments(youtube, video_id)
                data.extend(video_comments)
                requests_made += 1
            except HttpError as e:
                error_message = json.loads(e.content)['error']['errors'][0]['reason']
                if error_message == 'commentsDisabled':
                    print(f"Comments are disabled for video ID {video_id}. Skipping...")
                elif error_message == 'quotaExceeded':
                    print(f"Quota exceeded while retrieving comments for video ID {video_id}. Returning available data.")
                    return data
                elif error_message == 'videoNotFound':
                    print(f"Video ID {video_id} not found. Skipping...")
                    continue  # Skip to the next video
                else:
                    # Handle other types of errors if needed
                    print(f"Error occurred while fetching comments for video ID {video_id}: {error_message}")

    return data

def retrieve_all_comments(youtube, video_id):
    data = []

    # Retrieve comments for the video
    next_page_token = None
    while True:
        try:
            video_comments = get_comments_for_video(youtube, video_id, page_token=next_page_token)
            data.extend(video_comments)
            next_page_token = None
            for comment in video_comments:
                if 'nextPageToken' in comment:
                    next_page_token = comment['nextPageToken']
                    break
            if not next_page_token:
                break
        except HttpError as e:
            error_message = json.loads(e.content)['error']['errors'][0]['reason']
            if error_message == 'commentsDisabled':
                print(f"Comments are disabled for video ID {video_id}. Skipping...")
                break
            elif error_message == 'quotaExceeded':
                print(f"Quota exceeded while retrieving comments for video ID {video_id}. Returning available data.")
                break
            elif error_message == 'videoNotFound':
                print(f"Video ID {video_id} not found. Skipping...")
                break
            else:
                # Handle other types of errors if needed
                print(f"Error occurred while fetching comments for video ID {video_id}: {error_message}")
                break

    return data

def get_comments_for_video(youtube, video_id, page_token=None):
    comments = []
    next_page_token = page_token

    while True:
        request = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=100,  # You can adjust this value as per your requirement
            pageToken=next_page_token
        )

        response = request.execute()

        for item in response["items"]:
            comment = item["snippet"]["topLevelComment"]["snippet"]
            # Convert timestamp to datetime object
            #timestamp = datetime.datetime.strptime(comment["publishedAt"], "%Y-%m-%dT%H:%M:%SZ")
            # Check if the comment is within the specified date range
            #if datetime.datetime(2018, 1, 1) <= timestamp <= datetime.datetime(2024, 1, 1):
            comments.append({
                    "comment_id": item["id"],
                    "video_id": video_id,
                    "video_title": get_video_title(youtube, video_id),
                    "text": comment["textDisplay"],
                    "timestamp": comment["publishedAt"],
                    "author": comment["authorDisplayName"],
                    "author_email": comment["authorChannelId"]["value"],
                    "reply_to": None  # Add a column to indicate it's a top-level comment
            })

        next_page_token = response.get("nextPageToken")
        if not next_page_token:
            break

    return comments

def get_video_title(youtube, video_id):
    request = youtube.videos().list(
        part="snippet",
        id=video_id
    )

    response = request.execute()
    return response["items"][0]["snippet"]["title"]
```

# Translate Singlish to Sinhala


```python
import pandas as pd
from googletrans import Translator


translator = Translator()

def translate_to_sinhala(text):
    try:
        translation = translator.translate(text, src='en', dest='si')
        return translation.text
    except Exception as e:
        print(f"Translation failed for text: {text}")
        return ''

df['sinhala_text'] = df['text'].apply(translate_to_sinhala)

print(df)
```
